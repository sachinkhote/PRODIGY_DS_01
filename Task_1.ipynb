```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
file_path = 'E-commerce Customer Behavior - Sheet1.csv'
data = pd.read_csv(file_path)
```

```python
data.head()
```

```python
print(data.describe())
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(16, 14))

# 1. Boxplot for each categorical variable
plt.subplot(2, 2, 1)
sns.boxplot(x='Gender', y='Total Spend', data=data, palette='Set2')
plt.title('Boxplot of Total Spend by Gender')

plt.subplot(2, 2, 2)
sns.boxplot(x='Membership Type', y='Total Spend', data=data, palette='Set2')
plt.title('Boxplot of Total Spend by Membership Type')

# 2. Violinplot for each categorical variable
plt.subplot(2, 2, 3)
sns.violinplot(x='City', y='Total Spend', data=data, palette='Set2')
plt.title('Violinplot of Total Spend by City')

plt.subplot(2, 2, 4)
sns.violinplot(x='Discount Applied', y='Total Spend', data=data, palette='Set2')
plt.title('Violinplot of Total Spend by Discount Applied')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 6))

# 3. Histogram for the 'Age' variable
plt.subplot(1, 2, 1)
sns.histplot(data['Age'], bins=10, kde=True, color='skyblue')
plt.title('Histogram of Age')

# 4. Pie Chart for 'Gender' distribution
plt.subplot(1, 2, 2)
gender_counts = data['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Gender Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 12))

# 5. Scatter Plot between 'Items Purchased' and 'Total Spend'
plt.subplot(2, 2, 1)
sns.scatterplot(x='Items Purchased', y='Total Spend', data=data, color='skyblue')
plt.title('Scatter Plot: Items Purchased vs Total Spend')

# 6. Line Chart for 'Days Since Last Purchase'
plt.subplot(2, 2, 2)
sns.lineplot(x='Days Since Last Purchase', y='Total Spend', data=data, color='green')
plt.title('Line Chart: Days Since Last Purchase vs Total Spend')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 12))
plt.subplot(3, 2, 1)
sns.boxplot(x='Membership Type', y='Total Spend', data=data, palette='Set2')
plt.title('Boxplot of Total Spend by Membership Type')

# 2. Violinplot for 'Total Spend' by 'Membership Type'
plt.subplot(3, 2, 2)
sns.violinplot(x='Membership Type', y='Total Spend', data=data, palette='Set2')
plt.title('Violinplot of Total Spend by Membership Type')

# Visualizations for Relationships (Scatter and Line)
plt.figure(figsize=(16, 12))

# 5. Scatter Plot between 'Items Purchased' and 'Total Spend' by 'Membership Type'
plt.subplot(2, 2, 1)
sns.scatterplot(x='Items Purchased', y='Total Spend', hue='Membership Type', data=data, palette='Set2')
plt.title('Scatter Plot: Items Purchased vs Total Spend by Membership Type')

# 6. Line Chart for 'Days Since Last Purchase' by 'Membership Type'
plt.subplot(2, 2, 2)
sns.lineplot(x='Days Since Last Purchase', y='Total Spend', hue='Membership Type', data=data, palette='Set2')
plt.title('Line Chart: Days Since Last Purchase vs Total Spend by Membership Type')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 18))

# 1. Boxplot for 'Total Spend' by 'Days Since Last Purchase'
plt.subplot(3, 2, 1)
sns.boxplot(x='Days Since Last Purchase',  data=data, palette='Set2')
plt.title('Boxplot of Total Spend by Days Since Last Purchase')

# 2. Violinplot for 'Total Spend' by 'Days Since Last Purchase'
plt.subplot(3, 2, 2)
sns.violinplot(x='Days Since Last Purchase',  data=data, palette='Set2')
plt.title('Violinplot of Total Spend by Days Since Last Purchase')

# 3. Histogram for 'Age' by 'Days Since Last Purchase'
plt.subplot(3, 2, 3)
sns.histplot(data['Age'], bins=10, kde=True, color='skyblue', label='All', alpha=0.7)
sns.histplot(data[data['Days Since Last Purchase'] <= 30]['Age'], bins=10, kde=True, color='orange', label='Last 30 days', alpha=0.7)
plt.title('Histogram of Age by Days Since Last Purchase')
plt.legend()
plt.tight_layout()
plt.show()
```
